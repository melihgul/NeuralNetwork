```python
import torch
torch.manual_seed(1)
X = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float)
```

```python
def tan_act_func(x):
  return (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x))  #Tanh aktivasyon fonksiyonu

def sigmoid_act_func(x):
  return 1 / (1 + torch.exp(-x))  #sigmoid aktivasyon fonksiyonu

import torch.nn as nn

class My_Model1(nn.Module):
    def __init__(self):
        super(My_Model1, self).__init__()
        self.hidden = nn.Linear(3, 50)   #hidden layerımız 3 feature 50 nöron  
        self.output = nn.Linear(50, 1)   #output layerımız 50 nörondan alıp 1 nörona gönderecek

    def forward(self, x):
        x = tan_act_func(self.hidden(x))  
        x = sigmoid_act_func(self.output(x))  
        return x

model = My_Model1() 
```

```python
print(model(X))
```
